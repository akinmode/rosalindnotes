# Consensus and Profile

```python
sample = open("010_cons.txt", "r").read().split(">")
del sample[0]
```

```python
sample = [i.split("\n", 1) for i in sample]
```

```python
sample = [[i[0], i[1].replace("\n", "")] for i in sample]
```

```python
sample
```

```python
def consensus(sample):
    code = {'0':'A', '1':'C', '2':'G', '3':'T'}
    strand = [i[1] for i in sample]
    c_matrix = []
    for i in range(int(len(strand[0]))):
        c_matrix.append([j[i] for j in strand])
    dict_matrix = [[i.count("A"), i.count("C"), i.count("G"), i.count("T")] for i in c_matrix]
    a_m = [code[str(i.index(max(i)))] for i in dict_matrix]
    f_matrix = []
    for i in range(int(len(dict_matrix[0]))):
        f_matrix.append(" ".join([str(j[i]) for j in dict_matrix]))
    
    return "".join(a_m), {code[str(i)]: f_matrix[i] for i in range(4)}
```

```python
consensus(sample)
```
